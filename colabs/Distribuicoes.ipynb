```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações gerais de estilo
sns.set_style("whitegrid")

# 1. Média, Mediana e Moda com Visualização

data1 = [2, 3, 3, 5, 7, 100, 10, 10, 12, 200]
mean1 = np.mean(data1)
median1 = np.median(data1)
#mode1 = stats.mode(data1).mode[0]

plt.figure(figsize=(8, 4))
sns.histplot(data1, bins=10, kde=True, color="skyblue", edgecolor="black")
plt.axvline(mean1, color='red', linestyle='dashed', linewidth=2, label=f'Média: {mean1:.2f}')
plt.axvline(median1, color='green', linestyle='dashed', linewidth=2, label=f'Mediana: {median1}')
#plt.axvline(mode1, color='blue', linestyle='dashed', linewidth=2, label=f'Moda: {mode1}')
plt.legend()
plt.title("Média, Mediana e Moda")
plt.show()
```

```python

# 2. Distribuição Normal e Intervalo de Confiança
mu, sigma = 50, 5
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma)

plt.figure(figsize=(8, 4))
plt.plot(x, y, color='purple', label="Distribuição Normal")
plt.fill_between(x, y, where=(x >= 45) & (x <= 55), color='gray', alpha=0.5, label="P(45 ≤ X ≤ 55)")
plt.axvline(45, color='black', linestyle='dashed')
plt.axvline(55, color='black', linestyle='dashed')
plt.legend()
plt.title("Distribuição Normal e Probabilidade")
plt.show()

```

```python

# 3. Distribuição Binomial
n, p = 5, 0.6
x_binom = np.arange(0, n+1)
y_binom = stats.binom.pmf(x_binom, n, p)

plt.figure(figsize=(8, 4))
plt.bar(x_binom, y_binom, color='orange', alpha=0.7)
plt.title("Distribuição Binomial - P(X=3)")
plt.xlabel("Número de Sucessos")
plt.ylabel("Probabilidade")
plt.show()

```

```python
# 4. Distribuição de Poisson
lambda_poisson = 4
x_poisson = np.arange(0, 10)
y_poisson = stats.poisson.pmf(x_poisson, lambda_poisson)

plt.figure(figsize=(8, 4))
plt.bar(x_poisson, y_poisson, color='blue', alpha=0.7)
plt.title("Distribuição de Poisson")
plt.xlabel("Número de Chegadas")
plt.ylabel("Probabilidade")
plt.show()

```

```python
# 5. Distribuição Exponencial
lambda_exp = 1
x_exp = np.linspace(0, 5, 100)
y_exp = lambda_exp * np.exp(-lambda_exp * x_exp)

plt.figure(figsize=(8, 4))
plt.plot(x_exp, y_exp, color='red', label="Distribuição Exponencial")
plt.fill_between(x_exp, 0, y_exp, alpha=0.3)
plt.title("Distribuição Exponencial - Tempo até o Próximo Evento")
plt.xlabel("Tempo")
plt.ylabel("Densidade de Probabilidade")
plt.legend()
plt.show()
```
